7. How do tuples ensure data integrity in Python?
    - Tuples help ensure data integrity in python because they are immutable. This means that once a tuple is created, its elements cannot be modified, added to, or removed.
8. What is a hash table, and how does it relate to dictionaries in Python?
    - A hash table is a data structure that stores key-value pairs and allows for very fast data retrieval. It works by using a hash function to compute an index (a hash code) for each key, which determines where the value is stored in memory.

10.  Explain why strings are immutable in Python?
    - Immutable strings greatly simplify memory allocation
    - if strings weren't immutable, you couldn't reliably use them as keys in a dict, since their hash value might suddenly change.
    - Python can optimize memory usage when it knows that strings won't change after they're created.
    - Immutability ensures that once a string is created, it can't be changed accidentally, which helps maintain data integrity.
11. What advantages do dictionaries offer over lists for certain tasks?

12. Describe a scenario where using a tuple would be preferable over a list?

13.  How do sets handle duplicate values in Python?
    - In Python, sets automatically handle duplicates by only allowing unique values. When you try to add a duplicate value to a set, it simply ignores the new value and doesn't add it again. This is because sets are designed to contain only one instance of each item.

15. Can you modify the elements of a tuple? Explain why or why not?
    - No, you cannot modify the elements of a tuple in Python. Tuples are immutable, which means that once they are created, their elements cannot be changed, added, or removed.
16. What is a nested dictionary, and give an example of its use case?
    - A nested dictionary is a dictionary where the values associated with some or all of the keys are themselves dictionaries. Essentially, it's a dictionary within a dictionary.
    - Let's say you need to store information about multiple students. For each student, you want to store their name, age, and a list of their courses. A nested dictionary can be used to represent this structure, where each student's details are stored as a dictionary, and all the student records are stored in a larger dictionary.
17. Describe the time complexity of accessing elements in a dictionary?
    - dictionaries are implemented using hash tables, which allows for very efficient access to elements. The time complexity of accessing elements in a dictionary is generally O(1) on average, meaning it takes constant time, regardless of the size of the dictionary.
18.  In what situations are lists preferred over dictionaries?
    - Lists are preferred over dictionaries in situations where order matters, where you need to store multiple duplicate values, or when you need to access elements by index rather than a key.
19. Why are dictionaries considered unordered, and how does that affect data retrieval?
    - Dictionaries in Python are considered unordered because they do not maintain the order in which elements are inserted. This is because dictionaries are implemented using hash tables, where the keys are hashed to a specific memory location, and the data is stored based on the hash value, not the insertion order.
20. Explain the difference between a list and a dictionary in terms of data retrieval.
    - Lists store elements in a sequential manner, meaning that each item is assigned an index starting from 0. To access a value in a list, you use the index of that item. Since the data is ordered, the index represents the position of the item in the list.
    - Dictionaries store data as key-value pairs, each element in a dictionary is associated with a key, and the value is accessed by referring to its corresponding key.

```python
# 13. Write a code to count how many times the number 2 appears in the tuple (1, 2, 3, 2, 4, 2)
t = (1, 2, 3, 2, 4, 2)
# Method 1 (using while loop)
n = len(t)
i = 0
count = 0
while i < n:
  if t[i] == 2:
    count +=1
  i+=1
print(count)
# Method 2 (using count function)
print(t.count(2))
```
